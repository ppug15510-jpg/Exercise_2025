```python
#154 파일의 특정 부분만 읽고 복사: f.seek()
with open('prophet_en.txt','r') as f:
  f.seek(335) #파일포인터를 335바이트 위치로 이동
  data = f.read()
print(data)
```

```python
#155 텍스트 파일 한 라인씩 읽기: f.readline(), f.readlines()
with open('prophet_en.txt','r') as f:
  line = f.readline()
  while line !='':
    print(f'>>{line[:-1]}')
    k = input('다음 줄: 엔터 입력. 종료:/q입력:')
    if k == '/q':
      break
    line = f.readline()
```

```python
with open('prophet_en.txt','r') as f:
  lines = f.readlines()
  for line in lines:
    print(f'>>{line[:-1]}')
    k = input('다음 줄: 엔터 입력. 종료:/q입력:')
    if k == '/q':
      break
```
