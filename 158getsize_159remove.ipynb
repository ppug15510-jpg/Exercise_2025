```python
#158 파일 크기 구하기: os.path.getsize()
from os.path import getsize

file1 = 'sunrise.jpg' #파일 이름만
file2 = 'C:/data/temp/DJI_0128.MP4' #절대경로
file3 = '../../temp/DJI_0129.MP4' #상대경로
```

```python
print(f'{file1}: {getsize(file1):,} 바이트')
print(f'{file2}: {getsize(file2):,} 바이트')
print(f'{file3}: {getsize(file3):,} 바이트')
```

```python
#159 파일 삭제하기: os.remove()
from os import remove

target = 'mydata.txt'
k = input(f'<{target}>을 삭제하시겠습니까? (y/n)')
if k.lower() == 'y':
  remove(target)
  print(f'<{target}>을 삭제했습니다.')
else:
  print(f'<{target}>을 삭제하지 않았습니다.')
```
