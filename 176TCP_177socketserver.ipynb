```python
#176 에코 클라이언트 만들기: socket
import socket

HOST = 'localhost'
PORT = 7777
def echoclient():
  with socket.socket(socket.AF_INET,socket.SOCK_STREAM) as sock:
    try:
      sock.connect((HOST,PORT))
      print(f'[{HOST}]와 연결되었습니다.')
      msg = input('메세지 입력:')
      sock.sendall(msg.encode())
      print(f'에코 서버로 전송한 데이터:{msg}')
      data = sock.recv(1024)
      print(f'에코 서버로부터 수신한 데이터:{data.decode()}')
    except Exception as e:
      print(e)

echoclient()
```

```python
#177 향상된 에코 서버 만들기: socketserver
import socketserver

HOST = ''
PORT = 7777

#이 클래스는 서버 하나당 단 한 번 초기화됩니다.
class MyTcpHandler(socketserver.BaseRequestHandler):
  #handle()에서 클라이언트 연결 처리를 위한 로직을 구현합니다.
  def handle(self):
    print(f'[{self.client_address[0]}] 연결됨')

    try:
      while True:
        self.data = self.request.recv(1024)
        if self.data.decode() == '/q':
          print(f'[{self.client_address[0]}] 사용자 종료')
          return
          print(f'수신 메시지 [{self.data.decode()}]를 에코 전송')
          self.request.sendall(self.data)
    except Exception as e:
        print(e)

def runServer():
  print('+++ 에코 서버를 시작합니다.')
  print('+++ 에코 서버를 끝내려면 Ctrl-c를 누르세요')

  try:
    server = socketserver.TCPServer((HOST,PORT),MyTcpHandler)
    server.serve_forever()
  except KeyboardInterrupt:
    print('--- 에코 서버를 종료합니다.')

runServer()
```
