```python
#160 파일 이름 변경/파일 이동: os.rename()
from os import rename

src = 'sunrise_copied.jpg'
dst = 'sunrise2.jpg'
rename(src, dst)
```

```python
#161 디렉토리에 존재하는 파일 목록 얻기: os.listdir(), glob.glob()
from os import listdir

folder = './' #현재 작업 디렉토리
ret = listdir(folder)
print(ret)
```

```python
from glob import glob

target = '*.txt'
ret = glob(target)
print(ret)
```

```python
target = '**'
ret = glob(target, recursive=True)
ret.sort()
for r in ret:
  print(r)
```
