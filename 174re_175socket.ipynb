```python
#174 정규표현식 응용
import re
p = re.compile(r'href=".*"',re.I)
```

```python
contents = """<a href="help.html">도움말을 보려면 여기를 클릭하세요</a><font size="15">"""
ret = p.match(contents)
print(ret)
```

```python
contents2 = """href="help.html">도움말을 보려면 여기를 클릭하세요</a><font size="15">"""
ret = p.match(contents2)
print(ret.group())
```

```python
ret = p.search(contents)
print(f'Match.group()={ret.group()}')
print(f'Match.start()={ret.start()}')
print(f'Match.end()={ret.end()}')
print(f'Match.span()={ret.span()}')
```

```python
ret = p.findall(contents)
print(ret)
```

```python
p = re.compile(r'src=\S+"',re.I)
with open('webpage.html') as f:
  contents = f.read()

imgurls = p.findall(contents)
for imgurl in imgurls:
  imgurl = imgurl.split('"')[1]
  print(imgurl)
```

```python
#175 에코 서버 만들기: socket
import socket

HOST = '' #빈 문자열이지만 서버가 구동되는 컴퓨터의 IP를 자동 할당
PORT = 7777

def echoServer():
  with socket.socket(socket.AF_INET,socket.SOCK_STREAM) as sock:
    sock.bind((HOST,PORT))
    sock.listen()
    print('클라이언트 연결을 기다리는 중...')
    conn,addr = sock.accept()
    with conn:
      print(f'{addr[0]}와 연결됨')
      while True:
        data = conn.recv(1024)
        if not data:
          break
          print(f'수신 메시지:{data.decode()}')
          conn.sendall(data)

echoServer
```
