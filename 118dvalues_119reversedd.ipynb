```python
#118 딕셔너리의 모든 값 조회하기: d.values()
greats = {'유비':89, '관우':95, '장비':91, '조운':94}
levels = greats.values()
print(list(levels))
```

```python
greats['조운'] = 97
print(list(levels))
```

```python
#119 딕셔너리에서 모든 키의 순서를 거꾸로 하기: reversed(d)
greats = {'유비':89, '관우':95, '장비':91, '조운':94}
levels = reversed(greats)
print(list(levels))
```
