```python
#166 파일인지 디렉토리인지 체크: os.path.isfile(), os.path.isdir()
from os import listdir
from os.path import isdir

fileList = []
dirList = []
files = listdir('./') #현재 작업 디렉토리의 파일 리스트
for file in files:
  if isdir(file):
    dirList.append(f'<DIR> {file}')
  else:
    fileList.append(file)

fileList.sort()
dirList.sort()

for d in dirList:
  print(d)

for f in fileList:
  print(f)
```
