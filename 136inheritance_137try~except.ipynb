```python
#136 클래스 상속
class Character:
  def __init__(self): #생성자에서 캐릭터의 공통 능력치를 1로 초기화
    self.strength = 1
    self.offense_power = 1
    self.intelligence = 1
    self.mentality = 1
    self.agility = 1
```

```python
class Wizard(Character):
  def setBasic(self, name=None, level=1, hp=100):
    self.name = name
    self.level = level
    self.hp = hp
```

```python
player1 = Wizard()
player1.setBasic('StormBaby')
print(f'[{player1.name}] 레벨: {player1.level}')
print(f'정신력: {player1.mentality}')
```

```python
class Wizard(Character):
  def __init__(self, name=None):
    super().__init__()
    self.name = name
    self.level = 1
    self.hp = 100
```

```python
player1 = Wizard('StormBaby')
print(f'[{player1.name}] 레벨: {player1.level}')
print(f'정신력: {player1.mentality}')
```

```python
#137 예외처리 1: try~except
fileList = ['file01.jpg','file02.jpg','file03','file04.jpg']
for file in fileList:
  fname = file.split('.') #파일 이름과 파일확장자 분리
  print(f'파일 이름: {fname[0]}, 파일확장자: {fname[1]}')
```

```python
for file in fileList:
  try:
    fname = file.split('.') #파일 이름과 파일확장자 분리
    print(f'파일 이름: {fname[0]}, 파일확장자: {fname[1]}')
  except:
    print(f'Error!! <{file}> 처리중 예외 발생!')
```
