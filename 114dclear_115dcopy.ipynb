```python
#114 딕셔너리의 모든 요소 삭제: d.clear()
greats = {'유비':89, '관우':95, '장비':91, '조운':94}
greats.clear()
print(greats)
```

```python
greats = {'유비':89, '관우':95, '장비':91, '조운':94}
del greats
print(greats)
```

```python
#115 딕셔너리의 모든 요소 복제: d.copy()
greats = {'유비':89, '관우':95, '장비':91, '조운':94}
buffer = greats.copy()
del buffer['관우']
print(f'greats:{greats}')
print(f'buffer:{buffer}')
```
